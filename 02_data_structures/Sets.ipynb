# SETS 

A set in Python is an unordered, mutable, and unique collection of elements. It
does not allow duplicate values

# Creating a Set 
 


```python
# Empty set (must use set(), not {})
empty_set = set()

# Set with elements
numbers = {1, 2, 3, 4, 5}

# Mixed data types
mixed_set = {1, "Hello", 3.14, True}

# Creating a set from a list
unique_numbers = set([1, 2, 2, 3, 4, 4, 5])

print(unique_numbers) # {1, 2, 3, 4, 5}
```

## Set Methods

```python
#add(x) -> Adds an element x to the set

s1= {1,2,3,4,5,6}
s1.add(7)
print(s1)
```

```python
# update(iterable) -> Adds multiple elements from an iterable

s1.update([8,9,10])
s1
```

```python
# remove(x) ->Removes  x from the set
#(raises an error if not found)

s1.remove(9)
s1
```

```python
# discard(x) ->Removes x from the set (does
# not raise an error if not found).

s1.discard(5)
s1
# s1.discard(18)
```

```python
# pop() ->Removes and returns a random element 

s1.pop()
```

```python
# clear() ->Removes all elements from the set 

s1.clear()

print(s1)
```

```python
# copy() ->Returns a shallow copy of the set
copy_set = s1.copy()
copy_set
```

# Set Operations

```python
# union(set2) -> Returns a new set with all unique elements from both sets.

s1={1,2,3,4,5}
s2={4,5,6,7,8}
s1.union(s2)
```

```python
# intersection(set2) ->Returns a set with elements common to both sets
s1.intersection(s2)
```

```python
# difference(set2) ->Returns a set with elements in set1 but not in set2. 
s1.difference(s2)
```

```python
# symmetric_difference(set2) -> Returns a set with elements in either set1 or set2 ,but not both .

s1.symmetric_difference(s2)
```

```python
# issubset(set2) ->Returns True if set1 is a subset of set2 .
s1={1,2,3,4,5,6}
s2={5,6}
s2.issubset(s1)
```

```python
# issuperset(set2) -> Returns True if set1 is a suuperset of set2.
s1.issuperset(s2)
```

# Key Properties of Sets:
- Unordered: No indexing or slicing.
- Unique Elements: Duplicates are automatically removed
- Mutable: You can add or remove elements.
