# LIST

- A list in Python is an ordered, mutable collection of elements. It can contain
elements of different types.

# Creating a list

```python
# Empty list
my_list = []

# List with elements
numbers = [1, 2, 3, 4, 5]
numbers

# Mixed data types
mixed_list =[1,"Hello",3.14,True]
mixed_list
```

# Common List Methods

```python
# append(x) -> adds an elements x to the end of a list

L1=[1,2,3,4,5]
L1.append(6)
L1
```

```python
# extend(iterable) -> Extends the list by appending all elements from an iterable.

L1.extend([8,9,7,0])
L1
```

```python
# insert(index, x) -> Inserts x at the specified index 

L1.insert(0,89)

L1
```

```python
# remove(x) -> removes the first occurence of x in the list

L1.remove(9)
L1
```

```python
# pop([index]) -> Removes and returns the element at index (last element if index is not provided)
L1.pop(0)
print(L1)
```

```python
# index(x) -> Returns the index of the first occurrence of x .

L1.index(5)
```

```python
# count(X) -> returns the number of times x appear in the list.
L1.count(5)
```

```python
# sort() -> sort the element in the ascending order.

L1.sort()
print(L1)
```

```python
# reverse() -> reverse the order of the list.

L1.reverse()
print(L1)
```

```python
# copy() -> returns the shallow copy of the list.

copy_list= L1.copy()
print(copy_list)
```

```python
#clear() -> removes all the element from the list.

L1.clear()
print(L1)
```
