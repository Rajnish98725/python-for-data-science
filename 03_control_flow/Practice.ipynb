# 2.Write a program to calculate grade of a student based on marks.

```python
# Assumption

# ≥90 → A

# ≥75 → B

# ≥60 → C

# ≥40 → D

# <40 → Fail

marks = 98

if marks >=90:
    print("A")
elif marks >=75:
    print("B")
elif marks >=60:
    print("C")
elif marks >=40:
    print("D")
else:
    print("Fail")
```

# 4. Write a program to check whether a character is an alphabet, digit, or special character.

```python
ch = input("Enter a character ")
if ch.isalpha():
    print("The character is alphabetic in nature")
elif ch.isdigit():
    print("The character is digit in nature")
else:
    print("The character is special character in nature")
```
